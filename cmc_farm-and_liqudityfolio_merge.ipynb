```python
import pandas as pd
```

```python
stable_coins = ['USDT', 'USDC', 'DAI', 'BUSD', 'SUSD']
```

```python
def get_stable_coin(pair1, pair2):
    if pair1 in stable_coins:
        return pair1
    elif pair2 in stable_coins:
        return pair2
    else:
        return'NA'
```

```python
farm_df = pd.read_csv("farming_data_cmc_new.csv", index_col=0)
liquidty_folio_df = pd.read_csv("uniswap_liquidityfolio_data.csv", index_col=0)
```

```python
farm_df = farm_df[['pair1','pair2','Pool_rate','Where','includes_stablecoin', 'Farm payment']]
```

```python
liquidty_folio_df = liquidty_folio_df[['pair1','pair2','Pool_rate','Where','includes_stablecoin', 'Farm payment']]
```

```python
liquidty_folio_df
```

```python
farm_df
```

```python
merged_df = farm_df.append(liquidty_folio_df)
merged_df
```

#### will filter out all the pairs from farm amd folio for symbol ETH

```python
defi_symbol = 'ETH'
```

```python
merged_eth_df = merged_df[(merged_df['pair1'] == defi_symbol) | (merged_df['pair2'] == defi_symbol)].copy().reset_index(drop=True)
```

```python
merged_eth_df
```

#### will extract data for 4 different scenario

### 1. [highest stabecoin pool rate, stablecoin symbol, where]

df[(df.includes_stablecoin == True) & (df.Farm_payment.isnull()) # has stablecoin and is a pool, not a farm since it doesn't contain a Farm payment

```python
scenario_1 = merged_eth_df[(merged_eth_df['includes_stablecoin'] == True ) & (merged_eth_df['Farm payment'].isnull())].sort_values(['Pool_rate']).iloc[-1]
stable_coin = get_stable_coin(scenario_1['pair1'], scenario_1['pair2'])
scenario_1_data = [scenario_1['Pool_rate'], stable_coin, scenario_1['Where']]
scenario_1_data
```

### [highest non-stablecoin pool rate, non-stablecoin symbol, where]
df[(df.includes_stablecoin == False) & (df.Farm_payment.isnull()) # does not have stablecoin and is a pool, not a farm since it doesn't contain a Farm payment

```python
scenario_2 = merged_eth_df[(merged_eth_df['includes_stablecoin'] == False ) & (merged_eth_df['Farm payment'].isnull())].sort_values(['Pool_rate']).iloc[-1]
stable_coin = get_stable_coin(scenario_2['pair1'], scenario_2['pair2'])
scenario_2_data = [scenario_2['Pool_rate'], stable_coin, scenario_2['Where']]
scenario_2_data
```

### [highest farm stablecoin rate, stablecoin symbol, where, farm payment]
df[(df.includes_stablecoin == True) & (df.Farm_payment is not None) # has stablecoin and is a farm, not a pool since it contain a Farm payment

```python
scenario_3 = merged_eth_df[(merged_eth_df['includes_stablecoin'] == True ) & (merged_eth_df['Farm payment'].notnull())].sort_values(['Pool_rate']).iloc[-1]
stable_coin = get_stable_coin(scenario_3['pair1'], scenario_3['pair2'])
scenario_3_data = [scenario_3['Pool_rate'], stable_coin, scenario_3['Where'], scenario_3['Farm payment']]
scenario_3_data
```

### [highest farm non-stablecoin rate, non-stablecoin symbol, where, farm payment]
df[(df.includes_stablecoin == False) & (df.Farm_payment is not None) # does not have stablecoin and is a farm, not a pool since it contain a Farm payment

```python
scenario_4 = merged_eth_df[(merged_eth_df['includes_stablecoin'] == False ) & (merged_eth_df['Farm payment'].notnull())].sort_values(['Pool_rate']).iloc[-1]
stable_coin = get_stable_coin(scenario_4['pair1'], scenario_4['pair2'])
scenario_4_data = [scenario_4['Pool_rate'], stable_coin, scenario_4['Where'], scenario_4['Farm payment']]
scenario_4_data
```
