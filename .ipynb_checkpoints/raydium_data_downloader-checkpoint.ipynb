```python
import requests
import pandas as pd
import pymysql
import numpy as np
```

```python
url = 'https://api.thegraph.com/subgraphs/name/pancakeswap/exchange'

post_form = {
    "operationName": "pairs",
    "variables": {
        "allPairs": ["a"]
    },
    "query": "fragment PairFields on Pair {\n  id\n  txCount\n  token0 {\n    id\n    symbol\n    name\n    totalLiquidity\n    derivedETH\n    __typename\n  }\n  token1 {\n    id\n    symbol\n    name\n    totalLiquidity\n    derivedETH\n    __typename\n  }\n  reserve0\n  reserve1\n  reserveUSD\n  totalSupply\n  trackedReserveETH\n  reserveETH\n  volumeUSD\n  untrackedVolumeUSD\n  token0Price\n  token1Price\n  createdAtTimestamp\n  __typename\n}\n\nquery pairs($allPairs: [Bytes]!) {\n  pairs( orderBy: trackedReserveETH, orderDirection: desc) {\n    ...PairFields\n    __typename\n  }\n}\n"
}
```

```python
??requests.post

```

```python
response = requests.post(url, json=post_form)
```

```python
json_data = response.json()
```

```python
pairs = json_data['data']['pairs']
```

```python
pairs[0]
```

```python
for pair in pairs:
    print([pair['id'], pair['token0']['symbol'], pair['token1']['symbol'], pair[], pair[], pair[], pair[], pair[]])
```
