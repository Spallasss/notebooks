```python
class Checks:
    
    def __init__(self,df_bars):
        _df_bars = df_bars[df_bars['datetime'].dt.time == datetime.time(16,0)].copy()
        _df_bars.reset_index(drop=True,inplace=True)
        _df_bars['closeHighestIn2days'] = _df_bars.Close.rolling(2).max()
        _df_bars['close5DaysAgo']= _df_bars.Close.rolling(5).apply(lambda x: x.iloc[0])
        _df_bars['closeHighestIn10Days'] = _df_bars.Close.rolling(10).max()
        self._df_bars = _df_bars.copy()
        
    def isTodayModayorFriday(self,today):
        day = today.strftime('%A')
        if day == 'Monday' or day == 'Friday':
            return True
        else:
            return False
        
    def entryConditionsCheck(self,today):
        df_bars = self._df_bars
        today_row = df_bars[df_bars['datetime'].dt.date == today.date()].copy()
        if len(today_row) == 0:
            return False
        today_row = today_row.iloc[0]
        cond1 = True if today_row['Close'] == today_row['closeHighestIn2days'] else False
        cond2 = True if today_row['Close'] > today_row['close5DaysAgo'] else False
        cond3 =  True if today_row['Close'] < today_row['closeHighestIn10Days'] else False
        return cond1 and cond2 and cond3
        
    def isEntryAllowed(self,today):
        check1 = self.isTodayModayorFriday(today)
        if check1:
            check2 = self.entryConditionsCheck(today)
            if check2:
                return True
        return False
```

```python
class OptionsRepository:
    
    def __init__(self,raw_opts):
        self._df_opts = raw_opts
    
    def getCallOptionsByDate(self,date):
        #date = datetime.date(2020,8,12)
        date_opts = self._df_opts[self._df_opts['date'].dt.date==date].copy()
        date_opts = date_opts[date_opts['call_put']=='C'].copy()
        return date_opts
    
    def filterOptionsByExpiry(self,date_opts,today_dt,max_days_left,least_days_req):
        #min_days_left
        expiry_opts = date_opts.copy()
        expiry_opts['days_left'] = (expiry_opts['option_expiration'] - today_dt).dt.days
        expiry_opts = expiry_opts[(expiry_opts['days_left'] < max_days_left) & (expiry_opts['days_left'] >= least_days_req)]
        return expiry_opts
    
    def filterOptionsByOutOfMOneyPer(self,expiry_opts,oom_per,oom_delta):
        oom_opts = expiry_opts.copy() 
        oom_opts['oom_per'] = ((oom_opts['strike'] / oom_opts['stock_price_for_iv'])-1)*100
        oom_opts =  oom_opts[oom_opts['oom_per']>0].copy() 
        oom_opts['delta_oom_per'] = abs(oom_opts['oom_per'] - oom_per)
        oom_opts = oom_opts[oom_opts['delta_oom_per']<=oom_delta].copy()
        oom_opts = oom_opts.sort_values(by=['delta_oom_per'])
        oom_opts.reset_index(drop=True,inplace=True)
        return oom_opts
    
    def getOptionsContract(self,today_date,max_days_left,least_days_req,oom_per,oom_delta):
        #today_date = datetime.date(2020,8,12)
        date_opts = self.getCallOptionsByDate(today_date)
        today_dt = datetime.datetime(today_date.year,today_date.month,today_date.day,0,0,0)
        expiry_opts = self.filterOptionsByExpiry(date_opts,today_dt,max_days_left,least_days_req)
        #display(expiry_opts)
        filtered_opts = self.filterOptionsByOutOfMOneyPer(expiry_opts,oom_per,oom_delta)
        #display(filtered_opts)
        if len(filtered_opts)>0:
            return filtered_opts.iloc[0]
        else:
            return []
    
    def getOptionsDataBySymbol(self,symbol):
        filtered_opt = self._df_opts[self._df_opts['option_symbol']==symbol].copy()
        filtered_opt.reset_index(drop=True,inplace=True)
        return filtered_opt
        
    def getOptionDataByDate(self,symbol,dt):
        opt_df =  self.getOptionsDataBySymbol(symbol)
        return opt_df[opt_df['date']==dt]
    
    def UpdateOptionContractAfterSplit(self,contract):
        exp = contract._expiration
        strike = contract._strike/5 
        opts = self._df_opts[self._df_opts['option_expiration'].dt.date==exp].copy()
        opts = opts[opts['strike']==strike].copy()
        if len(opts) > 0:
            contract._strike = strike
            contract._multiplier = 5*contract._multiplier
            contract._symbol = opts.iloc[0].option_symbol
        else:
            print('ERROR: NO after Split contract found')
        #return opts
    
    def UpdateOptionContractBeforeSplit(self,contract):
        exp = contract._expiration
        strike = contract._strike*5 
        opts = self._df_opts[self._df_opts['option_expiration'].dt.date==exp].copy()
        opts = opts[opts['strike']==strike].copy()
        opts = opts[opts['strike']==strike].copy()
        if len(opts) > 0:
            contract._strike = strike
            contract._multiplier = contract._multiplier/5
            contract._symbol = opts.iloc[0].option_symbol
        else:
            print('ERROR: NO before Split contract found')
        #return opts
        
    def contUpdateTest(self,contract):
        contract._strike = contract._strike*5
        contract._multiplier = contract._multiplier/5
        contract._symbol = 'AAA'
```

```python
class Contract:
    def __init__(self,ctype,symbol,expiry,multiplier,strike):
        self._type = ctype
        self._symbol = symbol
        self._expiration = expiry
        self._multiplier = multiplier
        self._strike = strike
        self._opt_entryPrice = 0.0
        self._premium_received = 0.0 
        self._stockPriceAtExpiry = 0.0
        
class Orderr:
    id = 0
    orderTypes =['MKT', 'LMT', 'STP-LOSS','STP-PFT']
    sides = ['BUY', 'SELL']
    def __init__(self,contract,orderType,side,qty,auxPrice=0):
        if orderType not in Orderr.orderTypes:
            raise ValueError("Order type is incorrect")
        if side not in Orderr.sides: 
            raise ValueError("side is incorrect")
        Orderr.id += 1
        self.Oid = Orderr.id
        self.contract = contract
        #self.symbol = symbol
        self.Otype =  orderType
        self.auxPrice = auxPrice
        self.side = side
        self.qty = qty 
        self.status = "created"
        self.lastUpated = None
        self.fillTime = None
        self.fillPrice = 0.0
        self.pos = 0
    
    def get_Oid(self):
        return self.Oid

class Broker:
    Portfolio = 0
    OpenOrders = []
    filledOrders = []
    CancelledOrders = []
    openContracts = []
    closedContracts = []
    position = {}
    allTrades = []
    currentDT = None
    EnableLogging = True
    logger = []
    
    def __init__(self,cash,isLogEnabled,optionsRepo):
        Broker.Portfolio =  cash
        Broker.EnableLogging = isLogEnabled
        self._optRep = optionsRepo
            
    def getPortfolio(self):
        return Broker.Portfolio
    
    def placeOrder(self,order):
        order.status = "Submitted"
        Broker.OpenOrders.append(order)
        if order.contract._expiration == Broker.currentDT:
            print('Can not place order of contract at its expiration day')
            self.cancelOrder(order)
        else:
            self.UpdateBroker(Broker.currentDT)
        return order
    
    def cancelOrder(self,order):
        if order.status == "Submitted":
            Broker.OpenOrders.remove(order)
            order.status = "Cancelled"
            Broker.CancelledOrders.append(order)
        elif order.status == "Cancelled":
            print("order cancelled before, can not be cancelled again")
        else:
            print("order filled , can not be cancelled")
            
    def getOrderById(self,oid):      
        for a in Broker.OpenOrders:
            if a.Oid == oid:
                return a
        for b in Broker.filledOrders:
            if b.Oid == oid:
                return b
        return None
    
    def getPosition(self):
        return Broker.position 

    def getTotalPnL(self):
        fills_df = self.getFilledOrdersdf()
        #fills_df['FillTime'] = pd.DatetimeIndex(fills_df['FillTime'])
        fills_df['pnl'] =0.0
        for index, row in fills_df.iterrows():
            val =  row["FillPrice"] * row["Qty"]
            fills_df.at[index,'pnl'] = val if row["Side"] == "SELL" else -1*val
        return fills_df.pnl.sum()
           
    def getClosedContractsdf(self):
        cc_df = pd.DataFrame(columns=['Type','Symbol','Expiration','Multiplier','Strike','Opt_EntryPrice','Premium_received','StockPriceAtExpiry']) 
        for cc in Broker.closedContracts:
            cc_df.at[len(cc_df)] = [cc._type,cc._symbol,cc._expiration,cc._multiplier,cc._strike,cc._opt_entryPrice,
                                    cc._premium_received,cc._stockPriceAtExpiry]
        return cc_df
    
    def getOpenOrdersdf(self):
        ods = pd.DataFrame(columns=['orderId','Symbol','Type','Side','Status','Qty','LastUpadted'])
        for od in Broker.OpenOrders:
            ods.at[len(ods)] = [od.Oid,od.contract._symbol,od.Otype,od.side,od.status,od.qty,od.lastUpated]
        return ods
    
    def getFilledOrdersdf(self):
        ofs = pd.DataFrame(columns=['orderId','Symbol','Type','Side','Status','Qty','FillPrice','FillTime','Pos'])
        for of in Broker.filledOrders:
            ofs.at[len(ofs)] = [of.Oid,of.contract._symbol,of.Otype,of.side,of.status,of.qty,of.fillPrice,of.fillTime,of.pos]
        return ofs

    def getCurrentBidAskClose(self,symbol,dt,side):
        curr_row = self._optRep.getOptionDataByDate(symbol,dt)
        # otions pricce 
        if side == 'bid':
            return curr_row.bid.values[0]
        elif side == 'ask':
            return curr_row.ask.values[0]
        elif side == 'stockclose':
            return curr_row.stock_price_for_iv.values[0]
        
    def updateOptContOnSplit(self):
        for cont in Broker.openContracts:
            self._optRep.UpdateOptionContractAfterSplit(cont)
    
    def UpdateBroker(self,dt):
        Broker.currentDT = dt
        fills = []
        if dt.date() == _split_dt.date():
            self.updateOptContOnSplit()
            
        for od in Broker.OpenOrders:
            symb = od.contract._symbol
            side = 'bid' if od.side=='SELL' else 'ask'
            current_price = self.getCurrentBidAskClose(symb,dt,side)
            
            if current_price == 0:
                self.cancelOrder(od)
                print('Order cancelled: Not valid price available in data')
                continue
            
            if od.status == "Submitted":
                if od.Otype == 'MKT':
                    od.status = "Filled"
                    od.fillTime = dt
                    fillPrice = current_price*od.contract._multiplier
                    od.contract._opt_entryPrice = current_price
                    od.fillPrice = fillPrice
                    poss = Broker.position[symb] if symb in Broker.position.keys() else 0
                    Broker.position[symb] = poss + od.qty if od.side=='BUY' else poss - od.qty
                    od.pos = Broker.position[symb]
                    premium =  od.qty*fillPrice
                    Broker.Portfolio =  Broker.Portfolio - premium if od.side == 'BUY' else Broker.Portfolio + premium
                    od.contract._premium_received = premium
                    fills.append(od)
                    if Broker.position[symb] != 0:
                        Broker.openContracts.append(od.contract)
                
        
        for of in fills:
            Broker.OpenOrders.remove(of)
            Broker.filledOrders.append(of)
            
        for cont in Broker.openContracts: # Options expiration (only short sell)
            if cont._expiration <= dt:
                stock_price = self.getCurrentBidAskClose(cont._symbol,dt,'stockclose')
                cont._stockPriceAtExpiry = stock_price
                if stock_price <= cont._strike:
                    Broker.position[cont._symbol] = 0
                    #
                else:
                    pos = Broker.position[cont._symbol]
                    #side = 'bid' if pos > 0 else 'ask'
                    #current_opt_price = self.getCurrentBidAskClose(dt,side)
                    Broker.Portfolio = Broker.Portfolio - abs(pos)*(cont._strike - stock_price)*cont._multiplier
                    Broker.position[cont._symbol] = 0
                
                Broker.closedContracts.append(cont)
                Broker.openContracts.remove(cont)
                
                    
    def ResetBroker():
        Broker.Portfolio = 0
        Broker.OpenOrders = []
        Broker.filledOrders = []
        Broker.CancelledOrders = []
        Broker.openContracts = []
        Broker.closedContracts = []
        Broker.position = {}
        Broker.allTrades = []
        Broker.currentDT = None
        Broker.currentClose = {}
        Broker.EnableLogging = True
        Broker.logger = []
        Orderr.id = 0  #to reset order ids (optional)
                
    def getEquityValue(self):
        e_val = 0
        for symb, pos in Broker.position.items():
            if pos != 0:
                e_val = e_val + pos*Broker.currentClose[symb]
        return Broker.Portfolio + e_val        
            
    def add_log(self,log_string):
        s = f'{next(iter(Broker.currentDT.values()))} : {log_string}'
        Broker.logger.append(s)

    def get_logs(self):
        logs_str = ""
        for s in Broker.logger:
            logs_str = logs_str + s +"\n"
        return logs_str
```

```python
# Running backtest for various out of money percentages
def run_backtest(raw_opts,ch,opts_repo,enableEntryCheck):
    df_Summary = pd.DataFrame(columns=['Cash','OutOfMoneyPer','MaxExpireDays','ProfitTrades',
                                       'LossTrades','TotalTrades','Profit','Loss','NetPnL'])
    
    for oom_per in range(5,100,5):
        print(f'Running Backtest for Out Of Money = {oom_per}%')
        cash = 1000000
        Broker.ResetBroker()
        OB = Broker(cash,True,opts_repo)

        for today in raw_opts.date.unique():    
            today = pd.to_datetime(today)
            #print(today.date())
            OB.UpdateBroker(today)
            pos = OB.getPosition()
            #print(pos)
            #isDateInRange = isDateInRangeOfEarnDates(today)
            is_entry_allowed = True
            if enableEntryCheck:
                is_entry_allowed = ch.isEntryAllowed(today)
            if is_entry_allowed and (len(pos)==0 or sum(pos.values()) == 0): #Entry
                dc = opts_repo.getOptionsContract(today.date(),_max_days_in_expiry,_least_days_in_expiry,oom_per,_allowed_oom_delta) #dummy contract
                if len(dc)==0:
                    continue
                #df = opts_repo.getOptionsDataBySymbol(dc.option_symbol)
                #print(dc)
                #print(df[['date','stock_price_close','bid','ask','close','strike','option_expiration']])
                contract = Contract('OPT',dc.option_symbol,dc.option_expiration,_opts_cont_multiplier,dc.strike)
                #print(contract)
                order = Orderr(contract,'MKT','SELL',1)
                OB.placeOrder(order)

                #break
            #print(OB.Portfolio)
        df_cc = OB.getClosedContractsdf()
        df_cc['pnl'] = 0.0
        df_cc['pnl_per'] = 0.0

        for i,r in df_cc.iterrows():
            pnl = 0.0
            if r['StockPriceAtExpiry'] <= r['Strike'] :
                pnl = r['Premium_received']
            else:
                pnl = r['Premium_received'] - ((r['StockPriceAtExpiry']-r['Strike'])*r['Multiplier'])
            df_cc.at[i,'pnl'] = pnl
            df_cc.at[i,'pnl_per'] = round((100*pnl)/r['Premium_received'],2)

        profitable_trades = len(df_cc[df_cc['pnl']>=0])
        loss_trades = len(df_cc[df_cc['pnl']<0])
        total_trades = len(df_cc)
        profit = round(df_cc[df_cc['pnl']>=0].pnl.sum(),2)
        loss = round(df_cc[df_cc['pnl']<0].pnl.sum(),2)
        net_pnl = round(df_cc.pnl.sum(),2)
        print(f'Total trades: {total_trades}')
        print(f'Profitable trades: {profitable_trades}')
        print(f'Loss trades: {loss_trades}')
        print(f'Profit: {profit}')
        print(f'Loss: {loss}')
        print(f'Net pnl: {net_pnl} \n')
        df_Summary.at[len(df_Summary)] = [cash,oom_per,_max_days_in_expiry,profitable_trades,loss_trades,total_trades,profit,loss,net_pnl]
    return df_Summary
```
