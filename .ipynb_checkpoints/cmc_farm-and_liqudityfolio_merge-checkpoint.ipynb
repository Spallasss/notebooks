```python
stable_coins = ['USDT', 'USDC', 'DAI', 'BUSD', 'SUSD']
```

```python
def get_stable_coin(pair1, pair2):
    if pair1 in stable_coins:
        return pair1
    elif pair2 in stable_coins:
        return pair2
    else:
        return'NA'
```
