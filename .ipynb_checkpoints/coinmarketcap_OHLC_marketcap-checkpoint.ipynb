### Method to fetch top n symbols/coin based on market_cap

```python
def fetch_top_n_market_cap_symbols_(n):
    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
    payload = {
        'convert':'USD',
        'aux':'cmc_rank',
        'limit':n
    }
    headers = {
      'Accepts': 'application/json',
      'X-CMC_PRO_API_KEY': api_key
    }
    
    data = requests.get(url, params=payload, headers=headers).json()['data']
    return data
```

### method to parse the top n market_cap symbols response ^^

```python
def parse_data(rows):
    parsed_data = []
    for data in rows:
        parsed_data.append([data['name'], data['symbol'],data['cmc_rank'], data['quote']['USD']['market_cap']])
    return parsed_data
```

### method to fetch OHLC for a passes start and end time. symols count should be frater than 1.

```python
def get_historical_data(start_time, end_time, symbols, parsed_data, count = 1000):
    url = "https://pro-api.coinmarketcap.com/v1/cryptocurrency/ohlcv/historical"
    payload = {
        'convert':'USD',
        'time_period':'hourly',
        'time_start':start_time,
        'time_end': end_time,
        'count':count,
        'interval':'hourly',
        'symbol':symbols
    }
    headers = {
      'Accepts': 'application/json',
      'X-CMC_PRO_API_KEY': api_key
    }
#     return requests.get(url, params=payload, headers=headers).json()
#     print(payload)
    try:
        response = requests.get(url, params=payload, headers=headers).json()
        data = response['data']
        get_historical_parsed_data(data, parsed_data, symbols)
        return data
    except Exception:
        print(traceback2.format_exc())
        print(response)
```

### method to parse the historical OHLC data

```python
def get_historical_parsed_data(data, parsed_data, symbol):
#     print(data)
    if True:
        if 'symbol' in data:
            symbol = data['symbol']
            quotes = data['quotes']
        else:
            quotes = data[symbol]['quotes']
        for row in quotes:
            time_str = row['time_close']
            o = row['quote']['USD']['open']
            h = row['quote']['USD']['high']
            l = row['quote']['USD']['low']
            c = row['quote']['USD']['close']
#             print([time_str, symbol, o, h, l, c])
            parsed_data.append([time_str, symbol, o, h, l, c])
    return parsed_data
```

### method to fetch data for pased symbols and start and end time. (a looped method for all the symbols). batching the symbols list.

```python
def get_data(start_time, end_time, symbols, parsed_data):
    symbols_size = len(symbols) if 100 > len(symbols) else len(symbols)
    i = 0
    batch_size = 1
    while i <  symbols_size:
        j = i + batch_size
        if j > symbols_size:
            j = symbols_size
        count = 10000
        s = ",".join(symbols[i:j])
        print(f'Index: {i}, Count:{count} Symbols: {s}')
        i = j
        data = get_historical_data(start_time, end_time, s, parsed_data, count)
        time.sleep(2)
#     print(data)
```

#### this method substract the benchmarked ETH 7 days pct change data to every coin's 7 days pct change

```python
def benchmark_with_eth(df, eth_df, symbols):
    columns = df.columns
    for symbol in symbols:
#         print(symbol)
        if symbol not in columns:
            continue
        tmp = df[symbol]
#         print(tmp)
        if not tmp.empty and symbol != 'ETH':
            df[symbol] = df[symbol] - eth_df
    return df
```

### coin_div: 

#### as in initial requirment :
coin_div -> benchmark (eth) -, where(eth -, pct_chg of coin, otherwise 0)

#### as in review:
use hourly pct change to calculate coin_div.  it should be 0 when eth hourly pct change is positive, and the pct change of the coin in hours when eth hourly pct change is negative

Based on above requirment below wethod will update hourly pct change for all the coins other that ETH as ETH is the bench mark here. and the logic is as below,

#### if eth hourly_pct_change is +ve, then the coin hourly_pct_change will be 0

#### if eht hourly_pct_change is -ve, then the coin hourly_pct_change will be the coin hourly_pct_change (no change)

```python
def update_coin_devi(df):
    df = df.copy()
    columns = df.columns
    for ind in df.index:
        for symbol in columns:
            if 'ETH' != symbol:
                df[symbol][ind] = 0 if df['ETH'][ind] >= 0 else df[symbol][ind]
    return df
```

```python
def get_next_trading_day_as_index(start_date, hour):
    return pd.to_datetime((start_date + timedelta(days = 1) ).replace(hour=hour, minute=59, second=0).strftime("%Y %m %d %H:%M:00"))
    
    
```

#### this method will ran whether the passed symbols is in top n or not on the given date

```python
def is_symbol_in_top_n(df, symbol, trade_date, max_rank):
    data = df.loc[trade_date]
    data = pd.DataFrame(data)
    data = data.rename(columns={data.columns[0]:'C'})
    todays_close_df = pd.DataFrame(data)
    todays_close_df = todays_close_df.sort_values(['C']).tail(max_rank)
    try:
        tmp = todays_close_df.loc[symbol]
        return True
    except:
        return False
    
    
```

### returns the close price of the symbol on given date and time.

```python
def get_close_price_on(df, symbol, trade_date):
    data = df.loc[trade_date]
    data = pd.DataFrame(data)
    data = data.rename(columns={data.columns[0]:'C'})
    todays_close_df = pd.DataFrame(data)
    try:
        return todays_close_df.loc[symbol]['C']
    except:
        return 0
```

```python
def trade_for_symbol(symbol):    
    #trade start date
    now = datetime.now()
    start_time = now - timedelta(days=3*29, hours=23, minutes = 55)
    start_balance = 10000
    trades = []
    #trading hour
    hour = 16
    trade_date = start_time + timedelta(days=7)
    trade_date = get_next_trading_day_as_index(trade_date, hour)
    positions = False
    # top n symbols based on the 7 days pct change
    top_7_days_pct_symbols_pct = int(1 * len(_7_days_pct_change_df.columns.values))
    top_negdic_score_symbols_pct = int(1 * len(_7_days_pct_change_df.columns.values))
    # max holding period
    max_hold = 6

    holding = 0
    trade_id = 1
    last_buy = [trade_date, 0, 0, 0, 0, start_balance]
    while trade_date.timestamp() < now.timestamp():
        is_symbol_in_top = is_symbol_in_top_n(_7_days_pct_change_df.copy(), symbol, trade_date, top_7_days_pct_symbols_pct) and is_symbol_in_top_n(neg_div_score.copy(), symbol, trade_date, top_negdic_score_symbols_pct) 
        todays_close = get_close_price_on(df.copy(), symbol, trade_date)
    #     print(f"trade_date:{trade_date} close:{todays_close} positions:{positions} is_symbol_in_top_n:{is_symbol_in_top_n} holding:{holding}")
        if todays_close <= 0:
            last_buy[0] = trade_date
            last_buy[2] = todays_close
            last_buy[1] = 0
            trades.append(last_buy)
        elif positions and ( not is_symbol_in_top or max_hold == holding):
            #Sell
#             print(f'{start_balance}  {last_buy[3]} ')
            start_balance = last_buy[5] + (last_buy[3] * ( todays_close - last_buy[2]))
            last_buy = [trade_date, 1, todays_close, last_buy[3], last_buy[4], start_balance]
#             print(last_buy)
            trades.append(last_buy)
            trade_id = trade_id + 1
#             positions = False
            #Buy
#             print(f'{start_balance/todays_close} {start_balance} {todays_close}')
            last_buy = [trade_date, 2, todays_close, start_balance/todays_close, trade_id, start_balance]
            trades.append(last_buy)
#             print(last_buy)
            holding = -1
            positions = True
        elif not positions and is_symbol_in_top:
            positions = True
            last_buy = [trade_date, 2, todays_close, start_balance/todays_close, trade_id, start_balance]
            trades.append(last_buy)
        elif len(last_buy):
            trades.append([trade_date, 0, todays_close, last_buy[3], trade_id, last_buy[3]*todays_close])
#             print(holding)
#             trades.append(last_buy)

        holding = holding + 1
#         print(last_buy)
#         print(f'{top_7_days_pct_symbols_pct} {top_negdic_score_symbols_pct}')
        trade_date = get_next_trading_day_as_index(trade_date, hour)
    trade_df = pd.DataFrame(trades, columns=['time','side','close','positions','trade_id', 'today_balance'])
#     if not trade_df.empty:
#         trade_df[['today_balance','close']].plot(title=symbol)
#     print(trade_df)
    return trade_df
```

#### return the top n coin symbols based on bassed pct_change or neg_div_score

```python
def get_todays_top_n_coin(df, trade_date,top_n):
    data = df.loc[trade_date]
    data = pd.DataFrame(data)
    data = data.rename(columns={data.columns[0]:'C'})
    todays_close_df = pd.DataFrame(data)
    todays_close_df = todays_close_df.sort_values(['C']).tail(top_n)
    return todays_close_df.index.values.tolist()
    
```

```python
def get_joins(symbols1, symbols2):
    symbols_list = symbols1.copy()
    for symbol in symbols2:
        if symbol not in symbols1:
            symbols_list.append(symbol)            
    return symbols_list
```

```python
def get_commons(symbols1, symbols2):
    symbols_list = []
    for symbol in symbols2:
        if symbol in symbols1:
            symbols_list.append(symbol)            
    return symbols_list
```

```python
def get_not_in_list(new_list, old_list):
    not_in_new = []
    for symbol in old_list:
        if symbol not in new_list:
            not_in_new.append(symbol)            
    return not_in_new
```

#### trading symbols

```python
def make_exit_symbol(df, trading_date, symbol, last_trade):
    quantity = last_trade[2]
    todays_close = get_close_price_on(df, symbol, trading_date)
    return quantity*todays_close
    
```

```python
def make_entry_symbol(df, trading_date, symbol, last_trade, fund):
    todays_close = get_close_price_on(df, symbol, trading_date)
    if not len(last_trade):
        last_trade = [trading_date,todays_close,fund/todays_close,fund]
        return last_trade
    last_trade[0] = trading_date
    last_trade[1] = todays_close
    last_trade[3] = last_trade[2]*todays_close
#     print(last_trade)
    return last_trade
    
```

```python
def get_todays_portfolio_balance(trades):
    balance = 0
    for symbol in trades:
        balance = balance + trades[symbol][3]
    return balance
```

### Portfolio Logic
1. on day 1, selectiong coins based on its 7 days pct change and negdiv_score. (top performer). And making entry of all the symbols.

2. next day again selecting symbols on same logic as aplied on day one. and perfoms the below actions
    
    1. hold the positions of the symbols if it was in last day trade
    2. exit the symbols who is not in top performer list today
    3. make entry for new symbols. (Funds will be trasfer to the new symbols by exiting the non performer symbols.)

3. then calculating the total balance for the day
    

by doing so, porfolio alwasy maintain same no of symbols each day.

```python
def plot_portfolio_performence(trade_date, fund, df, pct_df, neg_div_score, pct_changed_symbols_count, negdiv_score_symbols_count):    
    trades = {}
    last_day_trading_symbols = None
    is_first_day_trading = True
    portfolio_performence = []
    while trade_date.timestamp() < now.timestamp():
        s1 = get_todays_top_n_coin(_7_days_pct_change_df, trade_date, pct_changed_symbols_count)
        s2 = get_todays_top_n_coin(neg_div_score, trade_date, negdiv_score_symbols_count)
    #     print(f'{s1} {s2}')
        todays_trading_symbols = get_commons(s1, s2)
        if is_first_day_trading:
            print(fund * len(todays_trading_symbols))
            for symbol in todays_trading_symbols:
                trade = make_entry_symbol(df, trade_date, symbol, [], fund)
                trades[symbol] = trade
            is_first_day_trading = False
            last_day_trading_symbols = todays_trading_symbols.copy()
            continue
        tmp_trades = trades

        trades = {}

        todays_exit_symbols = get_not_in_list(todays_trading_symbols, last_day_trading_symbols)
        todays_common_symbols = get_commons(todays_trading_symbols, last_day_trading_symbols)
        todays_new_entry_symbols = get_not_in_list(todays_common_symbols, todays_trading_symbols)
    #     print(f'{}')
        new_symbol = []
        for symbol_c in todays_common_symbols:
            trade = tmp_trades[symbol_c]
            trade = make_entry_symbol(df, trade_date, symbol_c, trade, 0)
            trades[symbol_c] = trade
        n = len(todays_exit_symbols)
#         print(todays_exit_symbols)
#         print(todays_new_entry_symbols)
        exit_fund = 0
        for exit_symbol in todays_exit_symbols:
            exit_trade = tmp_trades[exit_symbol]
            fund = make_exit_symbol(df, trade_date, exit_symbol, exit_trade)
            exit_fund = exit_fund + fund
        if len(todays_new_entry_symbols):
            new_entry_fund = exit_fund/len(todays_new_entry_symbols)
            for entry_symbol in todays_new_entry_symbols:
                trade = make_entry_symbol(df, trade_date, entry_symbol, [], new_entry_fund)
                trades[entry_symbol] = trade
                new_symbol.append(entry_symbol)
#         for i in range(0,n):
#             exit_symbol = todays_exit_symbols[i]
#             entry_symbol = todays_new_entry_symbols[i]
#             exit_trade = tmp_trades[exit_symbol]
#             fund = make_exit_symbol(df, trade_date, exit_symbol, exit_trade)
#             trade = make_entry_symbol(df, trade_date, entry_symbol, [], fund)
#             trades[entry_symbol] = trade
#             new_symbol.append(entry_symbol)
        portfolio_performence.append([get_todays_portfolio_balance(trades), trade_date])
        trade_date = get_next_trading_day_as_index(trade_date, hour)
        last_day_trading_symbols = get_joins(todays_common_symbols, new_symbol)
    print(todays_trading_symbols)
    return portfolio_performence
```

```python
def plot_portfolio_for_config(days_pct_changed_symbols_pct, negdiv_score_symbols_pct, df, pct_df, neg_div_score_df):
    total_symbols_count = len(df.columns.values.tolist())
    pct_changed_symbols_count = int(total_symbols_count/100*days_pct_changed_symbols_pct)
    negdiv_score_symbols_count = int(total_symbols_count/100*negdiv_score_symbols_pct)
    
    now = datetime.now()
    start_time = now - timedelta(days=3*29, hours=23, minutes = 55)
    trades = {}
    #trading hour
    hour = 16
    trade_date = start_time + timedelta(days=0)
    trade_date = get_next_trading_day_as_index(trade_date, hour)
    print(trade_date)
    s1 = get_todays_top_n_coin(pct_df, trade_date, pct_changed_symbols_count)
    s2 = get_todays_top_n_coin(neg_div_score_df, trade_date, negdiv_score_symbols_count)
    symbols_to_be_trade_on_day_1 = get_commons(s1, s2)
#     print(f'Day 1 portfolio Symbols: \n{symbols_to_be_trade_on_day_1}')
#     print(f'Day 1 portfolio Symbols size: {len(symbols_to_be_trade_on_day_1)}')
    fund = 100000/len(symbols_to_be_trade_on_day_1)
    portfolio_performence = plot_portfolio_performence(trade_date, fund, df, pct_df, neg_div_score_df, pct_changed_symbols_count, negdiv_score_symbols_count)
    
    performance_df = pd.DataFrame(portfolio_performence, columns=['balance', 'time'])
    performance_df.index = pd.DatetimeIndex(performance_df['time'])
    del performance_df['time']
    performance_df.plot(figsize=[18,10])
    print(performance_df)
    
    
```
