```python
from lxml import html
import requests
import json
import pandas as pd
from datetime import datetime, date, timedelta
import numpy as np
import pymysql

def __connect__():
    db_user = 'coin'
    db_password = 'dragonstone#123'
    db_name = 'master_data'
    db_host = 'localhost'
    return pymysql.connect(
        unix_socket=None,
        host=db_host,
        user=db_user,
        password=db_password,
        db=db_name,
        charset='latin1',
        cursorclass=pymysql.cursors.DictCursor)


def get_cursor(connection):
    if connection is None:
        connection = __connect__()
    connection.ping(reconnect=True)
    return connection.cursor()

api_key = "e463c4036385474984f31cd59d529b29"
uniswap_v2_protocol = "uniswap_v2"
balancer_protocol = "balancer"


def get_protocal_data(protocol):
    response = requests.get('https://api.liquidityfolio.com/all_pools_returns', params={"api_key": api_key, "protocol": protocol})
    result = response.json()
    return result['data']


def parse_uniswap_pool_data(data):
    pairs = []
    for row in data:
        pair_address = row['pair_address']
        pair1 = row['pair_tokens'][0]['token_symbol']
        if pair1 == 'WETH':
            pair1 = 'ETH'

        pair2 = row['pair_tokens'][1]['token_symbol']
        if pair2 == 'WETH':
            pair2 = 'ETH'
        pair_name = f"{pair1}-{pair2}"

        ps_last_30_days = row['pair_trend']['pool_size']['last_30_days']
        ps_last_14_days = row['pair_trend']['pool_size']['last_14_days']
        ps_last_7_days = row['pair_trend']['pool_size']['last_7_days']
        ps_today = row['pair_trend']['pool_size']['today']

        dv_last_30_days = row['pair_trend']['daily_volume']['last_30_days']
        dv_last_14_days = row['pair_trend']['daily_volume']['last_14_days']
        dv_last_7_days = row['pair_trend']['daily_volume']['last_7_days']

        rp_last_30_days = row['pair_trend']['relative_price']['last_30_days']
        rp_last_14_days = row['pair_trend']['relative_price']['last_14_days']
        rp_last_7_days = row['pair_trend']['relative_price']['last_7_days']
        rp_today = row['pair_trend']['relative_price']['today']

        next_30_day_return_from_fees = row['return_forecast']['next_30_day_return_from_fees']
        next_30_day_return_from_uni = row['return_forecast']['next_30_day_return_from_uni']
        next_30_day_impermanent_loss = row['return_forecast']['next_30_day_impermanent_loss']
        next_30_day_total_return = row['return_forecast']['next_30_day_total_return']
        equivalent_apr = row['return_forecast']['equivalent_apr']

        pairs.append([pair_address, pair_name, ps_last_30_days, ps_last_14_days, ps_last_7_days, ps_today, dv_last_30_days, dv_last_14_days, dv_last_7_days, rp_last_30_days, rp_last_14_days, rp_last_7_days, rp_today, next_30_day_return_from_fees, next_30_day_return_from_uni, next_30_day_impermanent_loss, next_30_day_total_return, equivalent_apr])
    return pairs


def create_uniswap_pool_data_df(pairs):
    columns_str = 'pair_address, pool, ps_last_30_days, ps_last_14_days, ps_last_7_days, ps_today, dv_last_30_days, dv_last_14_days, dv_last_7_days, rp_last_30_days, rp_last_14_days, rp_last_7_days, rp_today, next_30_day_return_from_fees, next_30_day_return_from_uni, next_30_day_impermanent_loss, next_30_day_total_return, equivalent_apr'
    columns = columns_str.replace(' ', '').replace('ps_', 'pool_pair_size_').replace('dv_', 'pool_daily_volume_').replace('rp_', 'pool_relative_price_').split(',')
    df = pd.DataFrame(pairs, columns=columns)
    return df


def process_uniswap_pool_data_df(df):
    df[['pair1', 'pair2']] = df['pool'].str.split("-", expand=True)[[0, 1]]
    df['pool_rate'] = df['next_30_day_return_from_fees']
    df['farm_payment'] = np.nan
    df['where'] = 'Uniswap'
    df['percentage_yield_from_fees'] = np.nan
    df['includes_stablecoin'] = (df['pair1'].isin(stable_coins)) | (df['pair2'].isin(stable_coins))
    uniswap_v2_df = df[['pair_address', 'pair1', 'pair2', 'where', 'farm_payment', 'pool_rate', 'includes_stablecoin', 'percentage_yield_from_fees']]
    now = datetime.now()
    now = datetime(now.year, now.month, now.day)
    uniswap_v2_df['date'] = now 
    return uniswap_v2_df


def get_uniswap_pool_data_df():
    uniswa_data = get_protocal_data(uniswap_v2_protocol)
    pairs = parse_uniswap_pool_data(uniswa_data)
    df = create_uniswap_pool_data_df(pairs)
    df = process_uniswap_pool_data_df(df)
    return df


def parse_balancer_pool_data(data):
    pairs = []
    for row in data:
        pair_address = row['contract_address']
        pair1 = row['tokens'][0]['token_symbol']
        if pair1 == 'WETH':
            pair1 = 'ETH'
        pair2 = row['tokens'][1]['token_symbol']
        if pair2 == 'WETH':
            pair2 = 'ETH'

        Pool_rate = row['average_daily_pool_fees_last_30_days']
        includes_stablecoin = (pair1 in stable_coins or pair2 in stable_coins)
        percentage_yield_from_fees = row['percentage_yield_from_fees']
        where = 'balancer'
        Farm_payment = np.nan
        pairs.append([pair_address, pair1, pair2, where, Farm_payment, Pool_rate, includes_stablecoin, percentage_yield_from_fees])
    return pairs


def get_balancer_pool_data_df():
    balance_data = get_protocal_data(balancer_protocol)
    pairs = parse_balancer_pool_data(balance_data)
    balancer_df = pd.DataFrame(pairs, columns=['pair_address', 'pair1', 'pair2', 'where', 'farm_payment', 'pool_rate', 'includes_stablecoin', 'percentage_yield_from_fees'])
    now = datetime.now()
    now = datetime(now.year, now.month, now.day)
    balancer_df['date'] = now
    return balancer_df


def insert_liquidityfolio_pool_df(df):
    connection = __connect__()
    cursor = get_cursor(connection)
    df = df.astype(object).where(pd.notnull(df), None)
    for i, row in df.iterrows():
        dt = row["date"].timestamp()
        try:
            print(row["date"], row['pair_address'], row["pair1"], row["pair2"], row["pool_rate"], row["percentage_yield_from_fees"], row["farm_payment"], row["where"], row["includes_stablecoin"])
            sql = "INSERT INTO `liquidityfolio_pool` (`timestamp_dt`, `pair_address`, `pair1`,`pair2`,`pool_rate`,`percentage_yield_from_fees`,`farm_payment`,`where`,`includes_stablecoin`) VALUES (FROM_UNIXTIME(%s), %s, %s, %s, %s, %s, %s, %s, %s)"
            cursor.execute(sql, (dt, row['pair_address'], row["pair1"], row["pair2"], row["pool_rate"], row["percentage_yield_from_fees"], row["farm_payment"], row["where"], row["includes_stablecoin"]))
        except Exception as e:
            print(e)
    connection.commit()
    cursor.close()
    connection.close()


def update_liquidityfolio_pool_data():
    if date.today().weekday() == 0:
        print('downloading liquidity folio Data')
        uniswap_df = get_uniswap_pool_data_df(uniswap_data)
        insert_liquidityfolio_pool_df(uniswap_df)
        balancer_df = get_balancer_pool_data_df(balancer_data)
        insert_liquidityfolio_pool_df(balancer_df)
    else:
        print('skipping data download as its not monday')
```
