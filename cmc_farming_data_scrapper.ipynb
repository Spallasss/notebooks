```python
from lxml import html
import requests
import json
import pandas as pd
```

```python
stable_coins = "USDT, USDC, DAI, BUSD, SUSD, UST, TUSD".split(",")
stable_coins
```

```python
url = "https://coinmarketcap.com/yield-farming/"
page = requests.get(url)
tree = html.fromstring(page.content)
```

```python
data = tree.xpath('//script[@id="__NEXT_DATA__"]/text()')
```

```python
json_obj = json.loads(data[0])['props']['initialProps']['pageProps']
farming_list = json_obj['yieldFarmingList']
```

```python
farming_list
```

```python
farming_data = []
for item in farming_list:
    pool_list = item['poolList']
    for pool_item in pool_list:
        rewards = [f"{reward['symbol']}/{reward['name']}" for reward in pool_item['rewards']]
        farming_data.append([pool_item['name'], pool_item['pair'], pool_item['totalStake'], pool_item['yearlyROI'], pool_item['monthlyROI'], pool_item['dailyROI'], pool_item['impermanentLoss'], ",".join(rewards)])
        
```

```python
farming_data
```

```python
df = pd.DataFrame(farming_data, columns=['name','pair', 'tvl', 'APY_Yearly', 'APY_Monthly', 'APY_Daily', 'impermanent loass', 'rewards'])
```

```python
df
```

```python
df
```

```python
df['pair_size'] = df['pair'].str.split("-").apply(lambda x: len(x))
```

```python
df = df[df['pair_size'] <=2 ]
```

```python
df
```

```python
df[['pair1', 'pair2']] = df['pair'].str.split("-", expand=True)[[0,1]]
```

### cmc Farming symbols to modify (WBNB = BNB, vBNB = BNB, vETH = ETH, vUSDT = USDT, vBUSD = BUSD, vXVS = XVS, vSXP = SXP, vUSDC = USDC, vXRP = XRP, lowercase V followed by anything letters can remove the v and its the same)

```python
symbol_mappings = {}
symbol_mappings['WBNB'] = 'BNB'
symbol_mappings['vBNB'] = 'BNB'
symbol_mappings['vETH'] = 'ETH'
symbol_mappings['vUSDT'] = 'USDT'
symbol_mappings['vBUSD'] = 'BUSD'
symbol_mappings['vXVS'] = 'XVS'
symbol_mappings['vSXP'] = 'SXP'
symbol_mappings['vUSDC'] = 'USDC'
symbol_mappings['vXRP'] = 'XRP'
```

```python
for index, row in df.iterrows():
    if row['pair1'] in symbol_mappings:
        print(row['pair1'])
        df.loc[index, 'pair1'] = symbol_mappings[row['pair1']]
        row['pair1'] = symbol_mappings[row['pair1']]
    if row['pair2'] in symbol_mappings:
        print(row['pair2'])
        row['pair2'] = symbol_mappings[row['pair2']]
```

```python
df[df['pair2'] == 'BNB']
```

```python
df['Pool_rate'] = df['APY_Yearly']
df['Pool_size'] = df['tvl']
```

```python
df['includes_stablecoin'] = (df['pair1'].isin(stable_coins)) | (df['pair2'].isin(stable_coins))
```

```python
df['Farm payment'] = df['rewards'].str.split("/", expand=True)[0]
```

```python
df['Where'] = df['rewards'].str.split("/", expand=True)[1]
```

```python
df = df.reset_index(drop = True)
df
```
