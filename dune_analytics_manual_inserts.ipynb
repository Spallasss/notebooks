```python
def get_file_data_df(data_dir, file_name):
    column_name = file_name.replace(".csv", "")
    df = pd.read_csv(f"{data_dir}{file_name}")
    df.columns = ['date', column_name]
    return df
```

```python
def conver_files_to_one_df(data_dir):
    min_date = None
    join_df = None
    dfs = []
    files = os.listdir(data_dir)
    files.remove('.ipynb_checkpoints')
    for file_name in files:
        df = get_file_data_df(data_dir, file_name)
        least_index = df.date.min()
        if not min_date:
            min_date = least_index
            join_df = df
            continue
        if min_date > least_index:
            min_date = least_index
            dfs.append(join_df)
            join_df = df
        else:
            dfs.append(df)
    for df in dfs:
        join_df = pd.merge(join_df, df, how = 'left', left_on=['date'], right_on = ['date'])
    return join_df
    
```

```python
df = conver_files_to_one_df(data_dir)
```
